```python
import pandas as pd
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=50, n_features=2, centers=5)

df = pd.DataFrame()
df["x"] = x[:, 0]
df["y"] = x[:, 1]
df["category"] = y
```

```python
import matplotlib.pyplot as plt

for category in df['category'].unique():
    category_data = df[df['category'] == category]
    plt.scatter(category_data['x'], category_data['y'], label=f'Category {category}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=1.5,min_samples=3)
db.fit(df.iloc[:, [0, 1]])
df["cat_by_db"] = db.labels_
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import DBSCAN
import numpy as np

# Create a concentric circle dataset
X, _ = make_circles(n_samples=200, factor=0.5, noise=0.03)

# Apply DBSCAN to the dataset
dbscan = DBSCAN(eps=0.25, min_samples=5)
clusters = dbscan.fit_predict(X)

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title("DBSCAN Clustering of Concentric Circles")
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```
